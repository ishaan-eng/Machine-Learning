```python
from xgboost import XGBRegressor
from catboost import CatBoostRegressor

housing=pd.read_csv('Housing.csv')
X=housing.drop('price',axis=1)
y=housing['price']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=25)
ohe=OneHotEncoder(sparse_output=False,drop='first').set_output(transform='pandas')
trns=make_column_transformer((ohe,make_column_selector(dtype_include=object)),remainder="passthrough",verbose_feature_names_out=False).set_output(transform='pandas')
X_train_trnf=trns.fit_transform(X_train)
X_test_trnf=trns.transform(X_test)
rates=np.linspace(0.01,0.8,20)
n_est=[50,100,200]
depth=[3,5,None]
score=[]
for r in tqdm(rates):
    for d in depth:
        for n in n_est:
            gbm=XGBRegressor(max_depth=d,n_estimators=n,random_state=25,learning_rate=r,verbose=0)
            gbm.fit(X_train_trnf,y_train)
            y_pred=gbm.predict(X_test_trnf)
            score.append([d,n,r2_score(y_test,y_pred)])
df=pd.DataFrame(score,columns=["estimator","n_estimators","score"])
df.sort_values("score")
```
