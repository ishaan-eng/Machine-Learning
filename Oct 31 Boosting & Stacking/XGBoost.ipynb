```python

import pandas as pd
import numpy as np
from sklearn.linear_model import ElasticNet, LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_selector,make_column_transformer
from sklearn.metrics import classification_report, mean_absolute_error, r2_score, f1_score, accuracy_score,log_loss
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis,QuadraticDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.ensemble import VotingClassifier, VotingRegressor, BaggingClassifier, BaggingRegressor, \
    RandomForestClassifier, AdaBoostClassifier, AdaBoostRegressor,GradientBoostingClassifier,StackingClassifier
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from tqdm import tqdm
from xgboost import XGBClassifier
```

```python
sonar=pd.read_csv('sonar.csv')
y=sonar["Class"]
X=sonar.drop("Class",axis=1)
le=LabelEncoder()
y=le.fit_transform(y)
score=[]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=25)
l_r=np.linspace(0.01,0.8,20)
n_est=[50,100,200]
max_d=[3,5,None]
for e in n_est:
    for n in l_r:
        gbm=XGBClassifier(learning_rate=n,n_estimators=e,random_state=25)
        gbm.fit(X_train,y_train)
        y_pred=gbm.predict_proba(X_test)
        score.append([e,n,log_loss(y_test,y_pred)])

df=pd.DataFrame(score,columns=["estimator","learning rate","log loss"])
df.sort_values("log loss")
```

```python
from lightgbm import LGBMClassifier

rates=np.linspace(0.01,0.8,20)
n_est=[50,100,200]
depth=[3,5,None]
scores=[]
for r in tqdm(rates):
    for d in depth:
        for n in n_est:
            gbm=LGBMClassifier(max_depth=d,n_estmators=n,random_state=25,learning_rate=r,verbose=-1)
            gbm.fit(X_train,y_train)
            y_pred=gbm.predict_proba(X_test)
            score.append([d,n,log_loss(y_test,y_pred)])
df=pd.DataFrame(score,columns=["estimator","n_estimators","score"])
df.sort_values("score")
```
