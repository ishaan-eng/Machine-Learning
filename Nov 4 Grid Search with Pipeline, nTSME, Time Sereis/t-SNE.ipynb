```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
wisconson=pd.read_csv('BreastCancer.csv',index_col=0)
X=wisconson.drop('Class',axis=1)
y=wisconson['Class']
```

```python
prcomp=PCA().set_output(transform='pandas')
X_pca=prcomp.fit_transform(X)
pc_data=X_pca.copy()
pc_data['Class']=y
sns.scatterplot(data=pc_data,x='pca0',y='pca1',hue='Class')
plt.title('Using PCA')
plt.show()
```

```python
tsne=TSNE(random_state=25).set_output(transform="pandas")
X_tsne=tsne.fit_transform(X)
tsne_data=X_tsne.copy()
tsne_data['Class']=y
sns.scatterplot(data=tsne_data,x='tsne0',y='tsne1',hue='Class')
plt.title("Using TSNE")
plt.show()
```

Vehicle with Scaling

```python
vehicle=pd.read_csv('Vehicle.csv',index_col=0)
X=vehicle.drop('Class',axis=1)
y=vehicle['Class']
```

```python
scaler=StandardScaler().set_output(transform='pandas')
X_scl=scaler.fit_transform(X)
tsne=TSNE(random_state=25,perplexity=40).set_output(transform="pandas")
X_tsne=tsne.fit_transform(X_scl)
tsne_data=X_tsne.copy()
tsne_data['Class']=y
sns.scatterplot(data=tsne_data,x='tsne0',y='tsne1',hue='Class')
plt.title('Using t-SNE')
plt.show()
```

```python
prcomp=PCA().set_output(transform='pandas')
X_pca=prcomp.fit_transform(X_scl)
pc_data=X_pca.copy()
pc_data['Class']=y
sns.scatterplot(data=pc_data,x='pca0',y='pca1',hue='Class')
plt.title('Using PCA')
plt.show()
```
