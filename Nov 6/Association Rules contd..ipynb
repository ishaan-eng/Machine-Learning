```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules, apriori
import os
from ipywidgets import interact,widgets
from mlxtend.preprocessing import TransactionEncoder
```

```python
groceries=[]
with open("groceries.csv",'r') as f:groceries=f.read()
groceries=groceries.split('\n')
groceries_list=[]
for i in groceries:
    groceries_list.append(i.split(','))
```

```python
te=TransactionEncoder().set_output(transform='pandas')
te_ary=te.fit_transform(groceries_list)
itemsets=apriori(te_ary,min_support=0.002,use_colnames=True)
rules=association_rules(itemsets,metric='confidence',min_threshold=0.7)
rules=rules[['antecedents','consequents','support','confidence','lift']]
rules.sort_values(['lift','confidence'],ascending=[False,False])
```

```python
housing=pd.read_csv('Housing.csv')
```

```python
intervals=[(0,50000),(50000,100000),(100000,150000),(150000,200000)]
bins=pd.IntervalIndex.from_tuples(intervals)
housing['price_slab']=pd.cut(housing['price'],bins)
intervals=[(0,5000),(5000,10000),(10000,15000),(15000,20000)]
bins=pd.IntervalIndex.from_tuples(intervals)
housing['area_slab']=pd.cut(housing['lotsize'],bins)
housing.drop(['price','lotsize'],axis=1,inplace=True)
housing=housing.astype(object)
fp_df=pd.get_dummies(housing,prefix_sep='=')
fp_df
```

```python
itemsets=apriori(fp_df,min_support=0.2,use_colnames=True)
rules=association_rules(itemsets,metric='confidence',min_threshold=0.7)
rules=rules[['antecedents','consequents','support','confidence','lift']]
rules.sort_values(['lift','confidence'],ascending=[False,False])
```
