```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
from django.db.models.expressions import result
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.model_selection import KFold,cross_val_score, StratifiedKFold, GridSearchCV, RandomizedSearchCV,KFold
import numpy as np
import pickle
```

```python
winconsin=pd.read_csv('BreastCancer.csv',index_col=0)
y=winconsin['Class']
X=winconsin.drop('Class',axis=1)
```

```python
kfold=StratifiedKFold()
lr=LogisticRegression()
```

```python
params={'solver':['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga'],'C':np.linspace(0.001,4,20)}
gcv=GridSearchCV(lr,param_grid=params,cv=kfold,scoring='roc_auc')
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
df_cv=pd.DataFrame(gcv.cv_results_)
df_cv
```

```python
rgcv=RandomizedSearchCV(lr,param_distributions=params,cv=kfold,scoring='roc_auc',n_iter=20,random_state=25)
rgcv.fit(X,y)
```

```python
print(rgcv.best_params_)
print(rgcv.best_score_)
```

Grid Search using Random Forest Regressor

```python
rfr=RandomForestRegressor()
rfr.get_params()
```

```python
concrete=pd.read_csv('Concrete_Data.csv')
y=concrete['Strength']
X=concrete.drop('Strength',axis=1)
max_features=[3,4,5,6,7]
max_depth=[3,5,None]
min_samples_split=[2,5,10,20]
min_samples_leaf=[1,5,10,20]
params={'max_features':[3,4,5,6,7],
'max_depth':[3,5,None],
'min_samples_split':[2,5,10,20],
'min_samples_leaf':[1,5,10,20]}
kfold=KFold(n_splits=5,shuffle=True,random_state=25)
```

```python
gcv = GridSearchCV(rfr, param_grid=params, cv=kfold, scoring='r2',verbose=3)
gcv.fit(X, y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
tst=pd.read_csv('testConcrete.csv')
bm=gcv.best_estimator_
bm.predict(tst)
```
